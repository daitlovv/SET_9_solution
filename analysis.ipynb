```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('results.csv')

algorithms = ['quicksort', 'mergesort', 'string_quicksort', 'string_mergesort', 'msd_radix', 'msd_radix_switch']
types = ['random', 'reverse', 'nearly', 'prefix']

for t in types:
    plt.figure(figsize=(12, 6))
    for algo in algorithms:
        data = df[(df['type'] == t) & (df['algorithm'] == algo)]
        plt.plot(data['size'], data['time_us'], marker='o', label=algo)
    sizes = sorted(df[df['type'] == t]['size'].unique())
    sizes = np.array(sizes)
    plt.plot(sizes, sizes * np.log2(sizes) * 0.001, 'k--', label='O(n log n)')
    plt.title(f'Time on {t}')
    plt.xlabel('Size')
    plt.ylabel('Time (microseconds)')
    plt.legend()
    plt.grid()
    plt.tight_layout()
    plt.savefig(f'time_{t}.png', dpi=150)
    plt.show()

for t in types:
    plt.figure(figsize=(12, 6))
    for algo in algorithms:
        data = df[(df['type'] == t) & (df['algorithm'] == algo)]
        plt.plot(data['size'], data['comparisons'], marker='o', label=algo)
    sizes = sorted(df[df['type'] == t]['size'].unique())
    sizes = np.array(sizes)
    plt.plot(sizes, sizes * np.log2(sizes) * 10, 'k--', label='O(n log n)')
    plt.title(f'Comparisons on {t}')
    plt.xlabel('Size')
    plt.ylabel('Comparisons')
    plt.legend()
    plt.grid()
    plt.tight_layout()
    plt.savefig(f'comparisons_{t}.png', dpi=150)
    plt.show()
```
